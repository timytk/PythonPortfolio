**Question 3**: Consider a Fourier series approximation of a function with the Fourier series given as $$S(x,n) = \frac{4}{\pi} \sum_{m=1}^{n}\,\frac{1}{2\,m-1}\,\sin\left(\frac{(2\,m-1)\,2\,\pi\,x}{T}\right)$$ for a fixed range of values $x/T$, i.e. values in the interval `[0,1]`, with $T$ the function period and $n$ the order of approximation, i.e. how many terms in the Fourier series should be used.

**Task**: Use `np.linspace(from, to, points)` in order to create 200 $x/T$ values and plot all Fourier Series curves for the order parameters $n=(3, 11, 40)$ with appropriate axis labels and a title on a single figure. Also include a figure legend, describing your three curves. Observe how the Fourier series approximates the original step function better and better.

Therefore, **(a)** write a function `fourier(values, order)` which takes the $x/T$ values from above as first input and the order parameter as a second input. It should calculate the corresponding Fourier sum and return the array of Fourier values.

Then **(b)** write a main program which creates the $x/T$ values as described above and loops through the order parameters from above in order to plot the requested curves.

```python
# function first, then main script for plotting
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Fourier function calculates each x value's corresponding y value and adding the finite sum
def fourier(values, order):
    yvalues = []
    for i in values:
        sxnsum = 0
        for j in range(1, order + 1):
            sxnsum +=  (1 / (2*j - 1)) * np.sin((2*j - 1) * 2 * np.pi * i)
        sxnfin = (4 / np.pi) * sxnsum
        yvalues.append(sxnfin)
    return yvalues

# Generating plot values
xtvalues = np.linspace(start = 0, stop = 1, num = 200)
order3 = fourier(xtvalues, 3)
order11 = fourier(xtvalues, 11)
order40 = fourier(xtvalues, 40)

# Graph settings
plt.figure(figsize = (8, 5))
plt.title("Fourier Series Approximations")
plt.plot(xtvalues, order3, linewidth = 1.5, label = "n = 3")
plt.plot(xtvalues, order11, linewidth = 1.5, label = "n = 11")
plt.plot(xtvalues, order40, linewidth = 1.5, label = "n = 40")
plt.xlabel("x/T Values")
plt.ylabel("S(x, n)")
plt.legend()
plt.show()
```
