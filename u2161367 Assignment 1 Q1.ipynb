**Question 1**: Fit the Higgs mass - write a function `fitter(xvalues, data, init)`, then write a function `fitfunc(...)` that describes the combined background and signal model to fit the data, i.e. it should be called in the `fitter()` function. Note that the symbol `...` indicates that the argument to the function are your responsibility. Create two pictures: 

**(a)** plot the data with cross markers ('+' symbol) and the best fit curve as red line on the first plot and 

**(b)** draw the residual ratios with cross markers on the second plot where the residual ratio is defined as the difference between data and fitted model, divided by data for every data point.

The fit function must calculate first the background model with 3 parameters $$b(m)=A\,\exp{}\left( b_{1}\,(m - 105.5)+b_{2}\,(m-105.5)^{2}\right).$$ Then the signal is added to the background and its model is $$s(m) = \frac{R}{\sigma\sqrt{2\,\pi}}\,\exp{}\left(-\frac{(m-\mu{})^{2}}{2\sigma^{2}}\right),$$ featuring another 3 parameters for a total of six parameters to fit.

A set of 6 initial values for the fit are given. You could draw the data and the initial values, inserted into your function, on the same plot and convince yourself that these are close to the best in order to check your function.

The `xvalues` represent masses (in GeV units) of potential particles decaying into two photons in the ATLAS detector at CERN. Any peak would correspond to a real particle having decayed while a smooth background represents chance coincidences or simply noise. The peak to fit at a bit above 125 GeV led to the announcement of the discovery of the Higgs particle at CERN in 2012. The data itself is read off a histogram (in counts) hence the xvalues really are bin centres at the mass values as indicated.

```python
'''
Fit Higgs mass peak from data.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Input data, x-value array and initial parameter values:
# Use as given!
# Start value parameter definitions, see equations for s(m) and b(m).
# init[0] = mu
# init[1] = sigma
# init[2] = R
# init[3] = A
# init[4] = b1
# init[5] = b2
init = (125.8, 1.4, 470.0, 5000.0, -0.04, -1.5e-4)
xvalues = np.arange(start=105.5, stop=160.5, step=1)
data = np.array([4780, 4440, 4205, 4150, 3920, 3890, 3590, 3460, 3300, 3200, 3000, 
                 2950, 2830, 2700, 2620, 2610, 2510, 2280, 2330, 2345, 2300, 2190, 
                 2080, 1990, 1840, 1830, 1730, 1680, 1620, 1600, 1540, 1505, 1450, 
                 1410, 1380, 1380, 1250, 1230, 1220, 1110, 1110, 1080, 1055, 1050, 
                 940, 920, 950, 880, 870, 850, 800, 820, 810, 770, 760])

# YOUR CODE HERE
def fitfunc(xvalues, mu, sigma, R, A, b1, b2):
    bm = A * np.exp(b1 * (xvalues - 105.5) + b2 * (xvalues - 105.5) ** 2)
    sm = R / (sigma * np.sqrt(2 * np.pi)) * np.exp(- (xvalues - mu) ** 2 / (2 * sigma ** 2))
    return bm + sm

def fitter(xvalues, data, init):
    param, cov = curve_fit(fitfunc, xvalues, data, p0 = init)
    return param


#  (a)
bestfit = fitter(xvalues, data, init)
yvals = xvalues

plt.figure(figsize = (9, 5))
plt.title("Higgs Mass Fit Function")
plt.plot(xvalues, fitfunc(xvalues, *bestfit), color = "red", linewidth = 1, label = "Best Fit Curve")
plt.scatter(xvalues, data, color = "black", marker = "+", label = "data")
plt.xlabel("Mass (GeV)")
plt.ylabel("Decay Count")
plt.legend()
plt.show()

# part (b)
ratiopnts = (data - fitfunc(xvalues, *bestfit))/data
plt.figure(figsize = (9, 5))
plt.title("Residual Plots")
plt.scatter(xvalues, ratiopnts, color = "black", marker = "+", label = "Residual Points")
plt.xlabel("Mass (GeV)")
plt.ylabel("Residuals")
plt.legend()
plt.show()
```
