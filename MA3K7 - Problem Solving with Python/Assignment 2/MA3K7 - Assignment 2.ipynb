## Code to find possible matrices and their determinant

```python
# Function for copycat strategy only for even numbers
def copycatEven(num):
    matList = []
    for i in range(2 ** (num * int(num / 2))):
    # Using binary numbers to fill in matrix
        binRep = bin(i)[2:].zfill(num * int(num / 2))
        # Matrix that we copy and invert
        matTemp = np.array(list(map(int, binRep))).reshape(num, int(num / 2)) 
        matNew = np.zeros((num, num))
        matNew[:, ::2] = matTemp  # Copy original columns to even columns
        matNew[:, 1::2] = 1 - matTemp  # Invert and copy original columns to odd columns
        
        matList.append(matNew)

    return matList

# Listing valid matrices with the Determinant
def ccPrintEven(num):
    matrix = copycatEven(num)
    for mat in matrix[::43]:
        det = calDet(mat)
        print(mat)
        print("Determinant:", det, "\n")

# Prints 10 out of the 2^8 different possibilites
print("Printing 6 cases:")
ccPrintEven(4)
```

```python
# Checking if there are any non-zero determinants
def detCheckEven(num):
    test = 0
    matList = copycatEven(num)
    for mat in matList:
        check = calDet(mat)
        test += check
    return test

# Checking if all possible copycat strategies return a 0 determinant
print(detCheckEven(4))
print(detCheckEven(6))
```
