```python
# Helping print LaTeX code to print terms
def latexPrint(sequence):
    seqLen = 2 * sequence.size 
    for i in range(seqLen):
        if i/2 % 1 == 0:
            x = int(i/2)
            print(sequence[x], "" , end="")
        if i/2 % 1 != 0:
            print("\\ ", end="")
            
latexPrint(seq)
```

```python
# Creating a function to be used later
def bracelet(x0, x1):
    # Setting Initial Conditions
    seq = np.array([x0, x1])
    test = 0

    # Calculating first step
    x_n = seq[-2] + seq[-1]
    x_n = x_n % 10
    seq = np.append(seq, [x_n])

    # Loop until starting terms are regenerated
    while test == 0:
        # Check for starting terms
        if (seq[-2] != x0) | (seq[-1] != x1):
            x_n = seq[-2] + seq[-1]
            x_n = x_n % 10
            seq = np.append(seq, [x_n])
        # Finshes by printing bracelet and size
        else:
            test = 1
            return seq[0: -2], seq.size - 2
```

```python
# Check for unique sets
def checkUnique(x):
    checkSet = set()
    startPair = set()
    for i in range(x):
        for j in range(x):
            seq1, _ = bracelet(i, j)
            sortSeq = tuple(np.sort(seq1))
            if sortSeq not in checkSet:
                checkSet.add(sortSeq)
                startPair.add((i, j))
    print(startPair)

checkUnique(10)
```
